### Applying Label Encoder

```python
car_train_df_new
```

```python
new_cols = ['X0', 'X1', 'X2', 'X3', 'X4', 'X5', 'X6', 'X8']
```

```python
from sklearn.preprocessing import LabelEncoder

label_encoder = LabelEncoder()
```

```python
car_train_features = car_train_df_new.drop("y", axis = 1)
y = car_train_df_new["y"]
```

```python
car_train_label_encoded=car_train_features.apply(label_encoder.fit_transform)
car_train_label_encoded
```

### Applying Dimensionality Reduction

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(car_train_label_encoded,y, test_size = 0.25 , random_state = 21 )
```

```python
from sklearn.preprocessing import StandardScaler

sc = StandardScaler()

X_train = sc.fit_transform(X_train)
X_test = sc.fit_transform(X_test)
```

```python
from sklearn.decomposition import PCA 

pca = PCA()

X_train_pca = pca.fit_transform(X_train)
```

```python
print("The varaince of the resulting PC components are : ", pca.explained_variance_ratio_)
```

```python
print(X_train_pca.shape)
```

```python
pca2 = PCA(n_components = 120) 

X_train_pca2 = pca2.fit_transform(X_train)
X_test_pca2 = pca2.fit_transform(X_test)
```

```python
print(X_train_pca2.shape)
print(X_test_pca2.shape)
```
